```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('/Users/jannellypadilla/Documents/Natural_Science_Dataset.csv',index_col=0)
data
 
```

```python
df = data[['Country','Pollutant','MEASURE',
           'FLIGHT','Flight type','FREQUENCY','Frequency',
           'SOURCE','Source of emissions', 
           'TIME','Time','Value']]
df
```

```python
df= df[~df['Frequency'].isin(['Monthly','Quarterly'])]
df_grouped = df[['FLIGHT','Country','Value']].groupby(by=['FLIGHT','Country']).sum().reset_index(inplace=False)
df_top_n = df_grouped[df_grouped['FLIGHT']=='TOTAL'].nlargest(10, 'Value')
df_top_n
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mlpt
import seaborn as sns
%matplotlib inline
df_grouped = df_grouped[(df_grouped['FLIGHT'] != 'TOTAL') & (df_grouped['Country'].isin(df_top_n['Country'].unique()))].sort_values('Value', ascending=False)
sns.barplot(
    data=df_grouped,
    x='Country',
    y='Value',
    errorbar=None
)
sns.barplot(
    data=df_grouped[df_grouped['FLIGHT']=='C'],
    x='Country',
    y='Value',
    errorbar=None
)

flight_p = mlpt.Patch(color='blue', label='FLIGHT - P')
flight_c = mlpt.Patch(color='orange', label='FLIGHT - C')
plt.legend(handles=[flight_p,flight_c])
plt.xticks(rotation=90)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Title")
plt.show()

```
